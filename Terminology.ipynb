```
import nltk
# for text analytics we use nltk
nltk.download()
# d --> all
```

```
import urllib.request

response = urllib.request.urlopen('https://www.tourismcares.org/mission')

html = response.read()
print(html)
```

```
from bs4 import BeautifulSoup #get rid of things like '//', webpage resource
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip = True)
print(text)
```

```
tokens  =[t for t in text.split()] #an indication of words
print(tokens)
```

```
freq = nltk.FreqDist(tokens)
for key,val in freq.items():
  print(str(key)+':'+str(val))
```

```
freq.plot(20,cumulative = False)
```

```
#use of stop-word
from nltk.corpus import stopwords
stopwords.words('english')
```

```
clean_tokens = tokens[:]
sr = stopwords.words('english')

for t in tokens:
  if t in stopwords.words('english'):
    clean_tokens.remove(t)
  elif t =='&':
    clean_tokens.remove(t)
```

```
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
  print(str(key)+':'+str(val))
```

```
freq.plot(20,cumulative = False)
```
